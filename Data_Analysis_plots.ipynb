## Check if age matters in marketing subscription for deposit

```python
agegroup = bank_df[(bank_df["y"]=='yes')]
```

```python
agegroup
```

```python
agegroup_count = pd.DataFrame(agegroup.groupby("age")["y"].count()).reset_index()
```

```python
agegroup_count
```

```python
plt.figure(figsize=(6,6))
plt.grid("darkgrid")
plt.scatter(agegroup_count["age"], agegroup_count["y"])
plt.xlabel("Age")
plt.ylabel("Subscription")
```

```python
## The graph displays how number of subscription changes over age. The teen age group has the least subscription and subscription increases as increasing age group
## the subscription reached at the peak in the age range from 30-35 and it slowly decreases with age.
```

## Check if marital status mattered for a subscription to deposit

```python
maritalgroup_count = pd.DataFrame(agegroup.groupby("marital")["y"].count()).reset_index()
```

```python
maritalgroup_count
```

```python
plt.figure(figsize=(6,6))
plt.grid("darkgrid")
plt.bar(maritalgroup_count["marital"], maritalgroup_count["y"], label="subscription")
plt.ylabel("subscription")
plt.xlabel("Marital")
plt.legend()
```

```python
## This graph shows that marital status influence the subscription and married people are amongest all are the highest susbscriber
```

## Check if age and marital status together mattered for a subscription to deposit scheme

```python
agemaritalgroup_count = pd.DataFrame(agegroup.groupby(["age", "marital"])["y"].count()).reset_index()
```

```python
agemaritalgroup_count
```

```python
plt.figure(figsize=(4,4))
plt.grid("whitegrid")
plt.bar(agemaritalgroup_count["marital"], agemaritalgroup_count["y"], label="y")
plt.bar(agemaritalgroup_count["marital"], agemaritalgroup_count["age"], label="age")
plt.ylabel("count")
plt.legend()
plt.figure(figsize=(4,4))
plt.grid("whitegrid")
plt.scatter(agemaritalgroup_count["age"], agemaritalgroup_count["y"])
plt.ylabel("count")
```

## Do feature engineering for the bank and find the right age effect on the campaign.

```python
bank_df["age"]>56
```

```python
conditions = [bank_df["age"] < 20, (bank_df["age"]>21) & (bank_df["age"]<=32), (bank_df["age"]>33) & (bank_df["age"]<=55), bank_df["age"]>56]
outputs = ["Teen", "Young", "Middle", "Old"]
```

```python
bank_df["age_category"] = np.select(conditions, outputs)
```

```python
bank_df
```

```python
bank_df
```

```python
target_campaign_age = bank_df.groupby("age_category")["campaign"].count()
```

```python
target_campaign_age 
```

```python
## Middle age people affect campaign more than any other age groups.
```
