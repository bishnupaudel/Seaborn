```python
# Showing different kind of plots using Seaborn
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
iris_bunch = load_iris()
```

```python
print(iris_bunch.DESCR)
```

```python
iris_df = pd.DataFrame(iris_bunch.data, columns = iris_bunch.feature_names)
```

```python
iris_df
```

```python
# Line plot
sns.set_style("whitegrid")
plt.figure(figsize=(5,5))
a = sns.lmplot(x="sepal length (cm)", y="sepal width (cm)", hue = "sepal width (cm)", data = iris_df, fit_reg=False)
```

```python
# Strip plot shows all the data with underlying distribution
sns.set_style('whitegrid')
plt.figure(figsize=(20,5))
sns.stripplot(x="sepal length (cm)", y = "sepal width (cm)", data = iris_df)
```

```python
# Swarm plot reflects distribution of the data showing all the data points
plt.figure(figsize=(20,5))
sns.swarmplot(x="sepal length (cm)", y = "sepal width (cm)", data = iris_df)
```

```python
# Bar plot are important for categorical data or to convert data into category
plt.figure(figsize=(20,5))
sns.barplot(x = "sepal length (cm)", y="sepal width (cm)", data = iris_df)
```

```python
# Count plot shows counts of each variable in a vertical bar 
plt.figure(figsize=(20,5))
sns.countplot(x = "sepal length (cm)", data = iris_df)
```

```python
# Point plot displays each point giver user guidance to follow the trend
plt.figure(figsize=(20,5))
sns.pointplot(x = "sepal length (cm)", y="sepal width (cm)", data = iris_df)
```

```python
# Boxplot is very useful in determing the outliers in the data 
plt.figure(figsize=(20,5))
sns.boxplot(x = "sepal length (cm)", y="sepal width (cm)", data = iris_df)
```

```python
# Violing plot that shows underlying distribution of the data with features present in the boxplot
sns.violinplot("sepal width (cm)", data = iris_df, dodge=True, vertical=False)
```

```python
# pair plot plots all the variables in a matrix 
sns.pairplot(iris_df)
```

```python
# Jointplot with Kernel Density Estimator
sns.jointplot("sepal length (cm)", "sepal width (cm)", data=iris_df, kind = "kde")
```

```python
# Making grid only
sns.JointGrid(x="sepal length (cm)", y="sepal width (cm)", data = iris_df)
```

```python
# Heatmap showing correlation among variables of iris data
sns.heatmap(iris_df.corr())
```

```python
# Distribution plot, data are normalized and fitted with undrlying distributions
sns.distplot(iris_df.iloc[:,3])
```

```python
# Box plot
sns.boxplot(iris_df.iloc[:,1], whis=0.5)
```
